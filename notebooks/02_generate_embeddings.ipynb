### 1: Import Libraries

```python
# Libraries for data handling and embeddings
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer

print("Libraries imported successfully!")

```

### 2: Load Cleaned Data

```python
# Load the cleaned tickets data
data = pd.read_csv("../data/interim/tickets_cleaned.csv")
data.head()
```

### 3: Choose Embedding Model

```python
# Initialize the SentenceTransformer model
model_name = "all-MiniLM-L6-v2"  # Small, fast, good quality embeddings
model = SentenceTransformer(model_name)

print(f"Model {model_name} loaded successfully!")
```

### 4: Generate Embeddings

```python
# Generate embeddings for each ticket text
embeddings = model.encode(data["clean_text"].tolist(), show_progress_bar=True)
print("Embeddings shape:", embeddings.shape)
```

### 5: Save Embeddings

```python
# Save embeddings for future use
np.save("../data/interim/ticket_embeddings.npy", embeddings)
print("Embeddings saved to ../data/interim/ticket_embeddings.npy")
```
